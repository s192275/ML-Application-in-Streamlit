```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV, cross_validate

best_models = {}
X_train, X_test, y_train, y_test = utils.tr_test_split(X, y, test_size = 0.4)
y = np.argmax(y)

xgboost_params = {"learning_rate": [0.1],
                  "max_depth": [4,5,6],
                  "n_estimators": [50,60,70],
                  "colsample_bytree": [0.5],
                  "objective": ['multi:softmax'],
                  "num_class":[8]}

clf = XGBClassifier().fit(X_train, y_train)
preds = clf.predict_proba(X_test)
roc_auc = roc_auc_score(y_test,preds, multi_class='ovr')
cv_results = cross_validate(clf, X, y, cv=5, scoring=roc_auc)
print(f"roc_auc (Before): {round(cv_results['test_score'].mean(), 4)}")
gs_best = GridSearchCV(clf, xgboost_params, cv=5, n_jobs=-1, verbose=False).fit(X, y)
final_model = clf.set_params(**gs_best.best_params_)

cv_results = cross_validate(final_model, X, y, cv=5, scoring=roc_auc)
print(f"roc_auc (After): {round(cv_results['test_score'].mean(), 4)}")
print(f"XGBoost best params: {gs_best.best_params_}", end="\n\n")
best_models["XGBoostClassifier"] = final_model
print(best_models)        
print(roc_auc)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix,classification_report
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier

y = diabetes['TARGET'] 
X = diabetes.drop(columns=['TARGET'], axis = 0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state= 42)
print("Test Size : 0.4")

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix')
    plt.show()

model = LGBMClassifier().fit(X_train, y_train)
y_pred = model.predict(X_test)
plot_confusion_matrix(y_test, y_pred)
print("LGBM - Test Size : 0.4")
print(classification_report(y_test, y_pred))


model = XGBClassifier().fit(X_train, y_train)
y_pred = model.predict(X_test)
plot_confusion_matrix(y_test, y_pred)
print("XGB - Test Size : 0.4")
print(classification_report(y_test, y_pred))

model = RandomForestClassifier().fit(X_train, y_train)
y_pred = model.predict(X_test)
plot_confusion_matrix(y_test, y_pred)
print("RF - Test Size : 0.4")
print(classification_report(y_test, y_pred))

model = LogisticRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)
plot_confusion_matrix(y_test, y_pred)
print("LR - Test Size : 0.4")
print(classification_report(y_test, y_pred))

model = DecisionTreeClassifier().fit(X_train, y_train)
y_pred = model.predict(X_test)
plot_confusion_matrix(y_test, y_pred)
print("DT - Test Size : 0.4")
print(classification_report(y_test, y_pred))
```
